```python
# Dependencies and Setup
import pandas as pd
import numpy as np


# File to Load 
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data_df = pd.read_csv(file_to_load)
purchase_data_df.head()
```

## Player Count

* Display the total number of players


```python
#Display the total number of players
players = str(purchase_data_df['SN'].nunique())
print ("There are " + players + " total players.")
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Calculate the number of unique items in the DataFrame
item_count = len(purchase_data_df["Item ID"].unique())

#Calculate the average price for items
average_price = round(purchase_data_df["Price"].mean(), 2)

#Calculate total number of purchases
purchases = purchase_data_df["Purchase ID"].count()

#Calculate total revenue
revenue = purchase_data_df["Price"].sum()

# Place all of the data found into a summary DataFrame
summary_table = []

summary_table.append(item_count)
summary_table.append("$" + str(average_price))
summary_table.append(purchases)
summary_table.append("$" + str(revenue))
pd.DataFrame([summary_table], columns = ["Number of Unique Items", "Average Purchase Price", "Total Number of Purchases", "Total Revenue"])

```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#Calculate count and percent by gender
Count = purchase_data_df.groupby('Gender').SN.nunique()
Total = purchase_data_df.SN.nunique()
Percent = round(Count/Total*100, 2).astype(str) + '%'

#Display the dataframe
Gender_df = pd.DataFrame({
    "Count": Count,
    "Percentage": Percent
})

Gender_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Remove space from Item Name column
Renamed_Purchase_Data_df = purchase_data_df.rename(
    columns={"Purchase ID": "Purchase_ID", "Item ID": "Item_ID", "Item Name": "Name"})

# Purchase Count
Purchase_Count = Renamed_Purchase_Data_df.groupby('Gender').Name.count()

# Average Purchase Price
Average_Purchase_Price = round(Renamed_Purchase_Data_df.groupby('Gender').Price.mean(), 2).map('${:,.2f}'.format)

Total_Purchase_Price = round(Renamed_Purchase_Data_df.groupby('Gender').Price.sum(), 2).map('${:,.2f}'.format)




GenderAnalysis_df = pd.DataFrame({
    "Purchase Count": Purchase_Count,
    "Average Purchase Price": Average_Purchase_Price,
    "Total Purchase Price": Total_Purchase_Price
})

GenderAnalysis_df

# Average Purchase Total per Person by Gender
TotalByPerson = Renamed_Purchase_Data_df.groupby(['Gender','SN']).Price.sum()
TotalByPerson_df = pd.DataFrame(TotalByPerson)
Average_Purchase_Price = round(TotalByPerson_df.groupby('Gender').Price.mean(), 2).map('${:,.2f}'.format)
Average_Purchase_Price_df = pd.DataFrame({
    "Average Purchase Price per Person": Average_Purchase_Price
})
Average_Purchase_Price_df

Analysis = pd.merge(GenderAnalysis_df, Average_Purchase_Price_df, on="Gender", how="inner")
Analysis
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
purchase_data_df
#Set your bin ranges and name your bins
bins = [0, 9, 14, 19, 24, 29, 34, 39, 40]
age_range = ["<10", "10-15", "15-20", "20-25", "25-30", "30-35", "35-40", "40+"]

#Cut your data
pd.cut(purchase_data_df["Age"], bins, labels=age_range).head()

purchase_data_df["Age Range"] = pd.cut(purchase_data_df["Age"], bins, labels=age_range)
purchase_data_df.head()

#Count by screen name and then calc percent
Count = purchase_data_df.groupby("Age Range").SN.nunique()
Total = purchase_data_df.SN.nunique()
Percent = round(Count/Total*100, 2).astype(str) + '%'

#Display the dataframe
Age_df = pd.DataFrame({
    "Count": Count,
    "Percentage": Percent
})

Age_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
Renamed_Purchase_Data_df
#Set your bin ranges and name your bins
bins = [0, 9, 14, 19, 24, 29, 34, 39, 40]
age_range = ["<10", "10-15", "15-20", "20-25", "25-30", "30-35", "35-40", "40+"]

#Cut your data
pd.cut(Renamed_Purchase_Data_df["Age"], bins, labels=age_range).head()

Renamed_Purchase_Data_df["Age Range"] = pd.cut(Renamed_Purchase_Data_df["Age"], bins, labels=age_range)
Renamed_Purchase_Data_df.head()

#Calculations by age range
PurchaseCountByAge = Renamed_Purchase_Data_df.groupby("Age Range").Purchase_ID.nunique()
AveragePurchasePriceByAge = round(Renamed_Purchase_Data_df.groupby('Age Range').Price.mean(), 2).map('${:,.2f}'.format)
TotalPurchasePriceByAge = round(Renamed_Purchase_Data_df.groupby('Age Range').Price.sum(), 2).map('${:,.2f}'.format)


#Display the dataframe
Age_df = pd.DataFrame({
    "Purchase Count": PurchaseCountByAge,
    "Average Purchase Price": AveragePurchasePriceByAge,
    "Total Purchase Value": TotalPurchasePriceByAge
})

Age_df


# Average Purchase Total per Person by Age Range
TotalByPersonByAge = Renamed_Purchase_Data_df.groupby(['Age Range','SN']).Price.sum()
TotalByPersonByAge_df = pd.DataFrame(TotalByPersonByAge)
Average_Purchase_Price_By_Age = round(TotalByPersonByAge_df.groupby('Age Range').Price.mean(), 2).map('${:,.2f}'.format)
Average_Purchase_Price_By_Age_df = pd.DataFrame({
    "Average Purchase Price per Person": Average_Purchase_Price_By_Age
})
Average_Purchase_Price_By_Age_df

AgeAnalysis = pd.merge(Age_df, Average_Purchase_Price_By_Age_df, on="Age Range", how="inner")
AgeAnalysis
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Create dataframe, perform group by and set variables
Popular_Item = Renamed_Purchase_Data_df.groupby(["Item_ID", "Name"])
Popular_Item_Count = Popular_Item["SN"].count()
Popular_item_PriceSum = Popular_Item["Price"].sum()
Popular_Item_Price = (Popular_item_PriceSum / Popular_Item_Count)
Popular_Item_Value = (Popular_Item_Price * Popular_Item_Count)

# Create new dataframe with the new variables
Pop_Item = pd.DataFrame({"Purchase Count": Popular_Item_Count,
                          "Item Price": Popular_Item_Price,
                          "Total Purchase Value": Popular_Item_Value})

# Sort, format, and display
Pop_Item = Pop_Item.sort_values("Purchase Count", ascending=False) 
Pop_Item["Item Price"] = Pop_Item["Item Price"].map("${:.2f}".format)
Pop_Item["Total Purchase Value"] = Pop_Item["Total Purchase Value"].map("${:.2f}".format)
Pop_Item = Pop_Item[["Purchase Count", "Item Price", "Total Purchase Value"]]
Pop_Item.head()
```
